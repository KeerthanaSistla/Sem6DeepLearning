```python
import numpy as np

x = np.array([[0,0], [0,1], [1,0], [1,1]])

y = []
n = 0
for i in range(2):
  for j in range(2):
    for k in range(2):
      for l in range(2):
        y.append(np.array([i, j, k, l]))
        n += 1

print("a. Howmany boolean functions can be designed using two binary inputs?")
print("Number of boolean functions that can be designed using two binary inputs is", n)

total = 2 ** (2 ** 2)
print("Inputs:", 2)
print("Total boolean functions:", 2)

def perceptron(x, y):
  w = np.random.rand(2)
  b = np.random.rand(1)
  for _ in range(100):
    error_count = 0
    for xi, yi in zip(x, y):
      z = np.dot(xi, w) + b
      y_pred = 1 if z >= 0.5 else 0
      error = yi - y_pred
      if error != 0:
        w += 0.1 * error * xi
        b += 0.1 * error
        error_count += 1
    if error_count == 0:
      return True
  return False

seperable = 0

print("\nb. For each Boolean function, determine weather it is linearly seperable or not")

for yi in y:
  print("Function:", yi)
  result = perceptron(x, yi)
  print("Is function linearly seperable?")
  print(result)
  if result:
    seperable += 1

print("\nc. Implement a single perceptron model and test whether it can correctly learn each boolean function.")
print("Number of linearly seperable functions is", seperable)
print("Number of non-linearly seperable functions is", len(y) - seperable)
print("It cant learn", len(y) - seperable, "because single perceptron is a linear model and those functions are not linearly separable")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.datasets import mnist

(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=25)

loss, accuracy = model.evaluate(X_test, y_test)

print("Model Accuracy =", accuracy)
```
